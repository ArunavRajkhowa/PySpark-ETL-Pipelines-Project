```python
from containers.etl_process import ETLProcess
```

```python
import os
def main():
    # Define input and output directories
    input_dir = "D:\\Coding Project\\PySpark-ETL-Pipelines-Project\\"
    output_dir = "D:\\Coding Project\\PySpark-ETL-Pipelines-Project\\"

    # List all CSV files in the input directory
    
    input_files = [file for file in os.listdir(input_dir) if file.endswith(".csv")]

    # Process each CSV file
    for file_name in input_files:
        input_path = os.path.join(input_dir, file_name)
        output_path = os.path.join(output_dir, file_name.split(".")[0] + "_transformed")

        # Create an instance of ETLProcess
        etl_process = ETLProcess(input_path, output_path)

        # Run the ETL process
        etl_process.run_etl()

if __name__ == "__main__":
    main()
```
